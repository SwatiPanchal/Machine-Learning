### Handling the noisy data and inconsistent data

```python
for i in data.columns:
    data[i] = data[i].astype(str)
    data[i] = data[i].str.replace(r'\W|\D|',"")
    data[i] = data[i].replace('', np.nan, regex=True)
    data[i] = data[i].astype(float,errors="ignore")
    data[i] = data[i].replace([np.inf, -np.inf], np.nan)
```

### Observing the unique data in each column after handling inconsistent and noisy data

```python
for i in data.columns:
    print(i)
    print(data[i].unique())
    print(data[i].unique().shape[0], data[i].shape[0], "\n")
```

### Missing values 

```python
for i in data.columns:
    print(i.ljust(20) , data[i].isnull().sum())
```

### Handling the Missing values using Random Sample Imputation

```python
def impute(df,col):
    for i in col:
        random_sample = df[i].dropna().sample(df[i].isnull().sum(),random_state=42)
        random_sample.index = df[df[i].isnull()].index
        df.loc[df[i].isnull(),i] = random_sample
```

```python
impute(data,data.columns)
```
